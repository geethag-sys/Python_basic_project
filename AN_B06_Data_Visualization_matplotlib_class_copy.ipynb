```python
import pandas as pd
import matplotlib.pyplot as plt


# ================================
# 1. LOAD DATA
# ================================

# Load clean data
df = pd.read_csv('https://raw.githubusercontent.com/GeethaGunasekaran1/Dataset_rep/refs/heads/main/Advertising.csv')
df
```

```python
df=df.drop("Unnamed: 0",axis=1)
df
```

```python
df.info()
```

```python


# ================================
# 2. CORRELATION HEATMAP
# ================================
correlation=df[['TV',"Radio","Newspaper","Sales"]].corr()
correlation


```

```python
plt.figure(figsize=(8,6))
plt.imshow(correlation,cmap="Blues")
plt.colorbar()
plt.title("Correlation Heatmap")
plt.xticks(range(len(correlation)),correlation.columns)
plt.yticks(range(len(correlation)),correlation.columns)
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.Heatmap(correlation,annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python

# ================================
# 3. SCATTER PLOTS (MULTIVARIATE ANALYSIS)
# ================================

# TV vs Sales
plt.figure(figsize=(8,4))
plt.title("Tv Budget Vs Sales")
plt.scatter(df['TV'],df["Sales"])
plt.xlabel("TV Budget")
plt.ylabel("Sales")
plt.show()
```

```python
# Radio vs Sales

plt.figure(figsize=(8,4))
plt.title('radio budget vs sales')
plt.scatter(df["Radio"],df['Sales'])
plt.xlabel('radio budget')
plt.ylabel('sales')
plt.show()
```

```python
# Newspaper vs Sales

plt.figure(figsize=(8,4))
plt.title('radio budget vs sales')
plt.scatter(df["Newspaper"],df['Sales'])
plt.xlabel('Newspaper budget')
plt.ylabel('sales')
plt.show()

```
